HEATMAP

```python
flight_data = pd.read_csv("Downloads/flight_delays.csv", index_col="Month")
```

```python
flight_data.head()
```

```python
# Visualization to see average delay for each airline by month
plt.figure(figsize=(14,7))
sns.heatmap(data=flight_data, annot=True)
```
